```python
iterations = 1000
learning_rate = 0.02
```

```python
# num_of_training_examples, num_of_features
m, n = X.shape
m, n
```


Screenshot from 2024-08-04 17-48-39.png

```python
# weight initialization
theta_1 = np.zeros(n)
theta_0 = 0
theta_1, theta_0
```

Screenshot from 2024-08-04 18-14-16.png

```python
J_history = []
theta0_history = []
theta1_history = []

# gradient descent
for i in range(iterations) :

  # Predicted output values
  Y_pred = X.dot(theta_1) + theta_0

  # Cost function (J)
  J =  np.sum(np.square(Y_pred - Y)) / (2*m)
  print ("iter / theta_0 / theta_1 /J ", i, theta_0, theta_1, J)

  # calculate gradients
  dtheta_1 = ((X.T).dot(Y_pred - Y)) / m
  dtheta_0 = np.sum(Y_pred-Y) / m

  # update parameters (weights)
  theta_1 = theta_1 - learning_rate * dtheta_1
  theta_0 = theta_0 - learning_rate * dtheta_0

  # store history values
  theta0_history.append(theta_0)
  theta1_history.append(theta_1)
  J_history.append(J)
```

```python
theta_1, theta_0
```

```python
fig, ax2 = plt.subplots()

ax2.plot(J_history, label='Loss function')
ax2.set_title(' $J(\\theta)$ vs number of iterations')
ax2.set_ylabel('Loss: $J(\\theta)$')

ax2.set_xlabel('Iterations')

fig.legend(loc='center');

# convert y-axis to Logarithmic scale
plt.yscale("log")
```
