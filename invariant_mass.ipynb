In this example we will learn how to calculate the invariant mass with the CMS open data. The invariant mass is an important concept for particle physicists when looking for new particles.

```python
# Import the necessary modules.

import pandas as pd
import numpy as np

dataset = pd.read_csv('http://opendata.cern.ch/record/307/files/Zmumu.csv')
# Credits to: https://opendata.cern.ch/record/307
```

```python
dataset.head()
```

```python
dataset.shape
```

Screenshot from 2024-10-02 13-05-26.png

```python
invariant_mass = np.sqrt(2*dataset.pt1*dataset.pt2*(np.cosh(dataset.eta1-dataset.eta2) - np.cos(dataset.phi1-dataset.phi2)))
```

```python
print(invariant_mass)
```

```python
import matplotlib.pyplot as plt

plt.hist(invariant_mass, bins=400)

# Name the axises and give the title.
plt.xlabel('Invariant mass [GeV]')
plt.ylabel('Number of events')
plt.title('The histogram of the invariant masses of two muons \n')
# Show the plot.
plt.show()
```

Z boson mass is around 91 GeV

```python
pt_muon1=dataset.pt1
pt_muon2=dataset.pt2

plt.hist(pt_muon1, bins=100, alpha=0.8, color='green', histtype='step')
plt.hist(pt_muon2, bins=100, alpha=0.8, color='red', histtype='step')

# Name the axises and give the title.
plt.xlabel('pT of muon [GeV]')
plt.ylabel('Number of events')
plt.title('The histogram of muon pt\n')
# Show the plot.
plt.show()
```
