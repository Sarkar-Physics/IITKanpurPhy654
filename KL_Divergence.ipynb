```python
import numpy as np
from matplotlib import pyplot as plt
```

Screen Shot 2024-09-30 at 12.48.08.png

```python
def normal_sample(x, mean, var):
    return ((1 / (np.sqrt(2 * np.pi * var)))
                * np.exp(-0.5 * (x - mean) ** 2 / var*var))
```

```python
# Setting mean, variance and x
mean = [0.0, 1.0]
var = [0.2, 0.4]
x_range = np.arange(-5.0, 5.0, 0.05)
curves = []
```

```python
# Computing distributions
for i in range(len(mean)):
    crv = [normal_sample(x, mean[i], var[i]) for x in x_range]
    curves.append(crv)
```

```python
# Visualize curves
plt.figure(figsize=(8, 4))

for i in range(len(curves)):
    plt.plot(x_range, curves[i],
             label=f"$\mu$={mean[i]}, $\sigma^2$={var[i]}")

plt.grid("on")
plt.legend()
plt.show()
```

```python
kl_loss = lambda mean_1, mean_2, var_1, var_2:(
            np.log((var_2 / var_1) ** 0.5)
              + (var_1 + (mean_1 - mean_2) ** 2) / (2 * var_2)
              - 0.5
           )

print("The KL Divergence between curve 0 and curve 1:")
print(f"{kl_loss(mean[0], mean[1], var[0], var[1]):.4f}")
```
